```python
import PyPDF2
import numpy as np
import pandas as pd
import re
import time
```

```python
import os
from pathlib import Path 
import chardet 
```

```python
root_dir = Path(".").resolve().parent
root_dir
```

```python
files_path=os.path .join(root_dir,"bioquimica_pd_sql","source") #,"raw","delitos_2020.csv")
files_path
#df = pd.read_csv(files_path,sep=",",dtype=str)
#df 
files = os.listdir(files_path)
```

```python
files_pdf_crime=[]
for file in files:
    if os.path.isfile(os.path.join(files_path, file)) and file.endswith('.PDF'):
        files_pdf_crime.append(file)
files_pdf_crime
```

```python
a = []
b = []
c = []
d = []
e = []
f = []
g = []
h = []
for file in files_pdf_crime:
    open_file = open('./source/'+file,'rb')
    pdf = PyPDF2.PdfReader(open_file)
    page = pdf.pages[0]
    text = page.extract_text()
    ###############################
    table = text.split('\n')
    ###############################
    for x in table[0:10]:
        matchcodigo = re.search("Código: ", x)
        if matchcodigo:
            codigo = x.split(': ')[1]
        matchdate= re.search("Informe: ", x)
        if matchdate:
            date =  x.split(': ')[1]
    for x in table:
        if ',' in x :
            x = x.replace(',', '.')
        match = re.search("\w+\s+\d+\.\d", x)
        if match :
            g.append(codigo)
            h.append(date)
            match3 = re.search("\w+-\w+\s+\d+\.\d", x)
            if match3:
                #print('grupo 3')
                data=x.split(' ')
                Componente = data[0] 
                ResultadoObtenido = data[1]
                Unidades = data[2]
                NivelComparación = data[3]
                ErrorRelativo = data[4]
                ZScore = data[5]
                a.append(Componente)
                b.append(ResultadoObtenido)
                c.append(Unidades)
                d.append(NivelComparación)
                e.append(ErrorRelativo)
                f.append(ZScore)
            else:
                print(x) 
                match1 = re.search("\w+\s+\w+\s+\d+\.\d", x)
                if match1:
                    print('grupo 1 o 2')            
                    data=x.split(' ')                
                    try:
                        if float(data[1]) : 
                                print('int' )  
                                print(float(data[1]))
                                Componente = data[0]
                                ResultadoObtenido = data[1]
                                Unidades = data[2]
                                NivelComparación = data[3]
                                ErrorRelativo = data[4]
                                ZScore = data[5]   
                                a.append(Componente)
                                b.append(ResultadoObtenido)
                                c.append(Unidades)
                                d.append(NivelComparación)
                                e.append(ErrorRelativo)
                                f.append(ZScore)                     
                    except:
                        if str(data[1]) :
                                print('str' )  
                                print( str(data[1]) )  
                                Componente = data[0] +' '+data[1]
                                ResultadoObtenido = data[2]
                                Unidades = data[3]
                                NivelComparación = data[4]
                                ErrorRelativo = data[5]
                                ZScore = data[6] 
                                a.append(Componente)
                                b.append(ResultadoObtenido)
                                c.append(Unidades)
                                d.append(NivelComparación)
                                e.append(ErrorRelativo)
                                f.append(ZScore)                       
```

```python
dataset = list(zip(a,b,c,d,e,f,g,h))
df = pd.DataFrame(dataset, columns=["Componente", "ResultadoObtenido", "Unidades","NivelComparación","ErrorRelativo","ZScore","codigo","fecha"])
```
