```python
import PyPDF2
import numpy as np
import pandas as pd
import re
import time
```

```python
import os
from pathlib import Path 
import chardet 
```

```python
root_dir = Path(".").resolve().parent
root_dir
```

```python
files_path=os.path .join(root_dir,"bioquimica_pd_sql","source") #,"raw","delitos_2020.csv")
files_path
#df = pd.read_csv(files_path,sep=",",dtype=str)
#df 
files = os.listdir(files_path)
```

```python
files_pdf_crime=[]
for file in files:
    if os.path.isfile(os.path.join(files_path, file)) and file.endswith('.PDF'):
        files_pdf_crime.append(file)
files_pdf_crime
```

```python
datos = []
meto = []
glob = []
medicamento = []
metodo = []
fecha = []
resultadoObtenido = []
for file in files_pdf_crime:
    open_file = open('./source/'+file,'rb')
    pdf = PyPDF2.PdfReader(open_file)
    number_pages = len(pdf.pages)    
    dimF = len(pdf.pages)
    for i in range( 2, dimF ):
        page = pdf.pages[i]
        text = page.extract_text() 
        ##########################################################
        index = 0
        table=text.split('\n')
        for x in table:
            if 'Resultado obtenido ' in x:
                pos = index
                pos
                break
            else:
                index = index +1
        res = table[12].split(' ')[2].replace(',', '.')
        fec = table[0].split(' ')[2]
        med = table[1]
        name_met = table[2].split(') ')[1]
        tabla = []
        for x in table[6:]:
            match = re.search("\d+\s+\d", x)
            if match:  
                tabla.append(x)
        tabla = tabla[:-3]  
        ##########################################################
        for x in tabla:
            aux = x.split(' ')
            i=1
            dato =  aux[0]
            while i < ( len(aux) -2 ) :
                dato = dato +' '+ aux[i]
                i=i+1
            met = aux[-2]
            glo =aux[-1]
            datos.append(dato)
            meto.append(met)
            glob.append(glo)
            medicamento.append(med)
            metodo.append(name_met)
            fecha.append(fec)
            resultadoObtenido.append(res)
```

```python
dataset = list(zip(datos,meto,glob,resultadoObtenido,medicamento,metodo,fecha))
dataset
producto = pd.DataFrame(dataset, columns=['datos','metodo','global','resultado obtenido','medicamento','metodo','fecha'])
producto.to_excel('Calidad.xlsx')
producto
```
